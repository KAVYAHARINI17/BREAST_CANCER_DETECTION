**IMPORTING THE LIBRARIES**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

**DATA IMPORTING AND PREPROCESSING**

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

**LOADING THE DATA INTO DATAFRAME AND PRINTING THE ROWS**

```python
df=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
df.head()
```

```python
df.tail()
```

**ADDING THE TARGET COLUMN**

```python
df['label']=breast_cancer_dataset.target
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#checking the distribution of target variables
df['label'].value_counts()
```

1-->BENIGN



0-->MALIGNANT

**SEPARATING THE INPUT COLUMS AND OUTPUT COLUMN**

```python
#axis 1 means to drop a column(vertical),axis 0 means to drop a row(horizontal)
X=df.drop(columns='label',axis=1)
Y=df['label']
```

```python
print(X)
```

```python
print(Y)
```

SPLITTING THE DATASET INTO TRAIN AND TEST

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.8,random_state=5)
print(X_train)
print(X_test)
```

**STANDARDIZE THE DATA**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train_std=scaler.fit_transform(X_train)
X_test_std= scaler.transform(X_test)
```

**BUILDING NEURAL NETWORKS**

IMPORTING TENSORFLOW AND KERAS LIBRARIES

```python
import tensorflow as tf
tf.random.set_seed(3) #to retreive the same output each time we run
from tensorflow import keras
```

```python
#setting up the layers of neural network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)), #input layer,converts multi dimensional input to 1D vector
    keras.layers.Dense(20,activation='relu'),  #hidden layer,A fully connected layer with 20 neurons and ReLU activations
    keras.layers.Dense(2,activation='sigmoid') #output layer,A fully connected layer with 2 neurons and sigmoid activation
])
```

```python
#compiling the NN
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#Training the NN
history = model.fit(X_train_std,Y_train,validation_split=0.1,epochs=10)
```

**VISUALIZATION**

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data','validation data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model accuracy loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data','validation data'],loc='lower right')
```

**ACCURACY OF THE MODEL OF TEST DATA**

```python
loss,accuracy = model.evaluate(X_test_std,Y_test)
print(accuracy)
```

```python
#PRINTING THE PREDICTION PROBABILITIES OF EACH CLASS FOR THE DATAPOINT

Y_pred = model.predict(X_test_std)
print(Y_pred)
```

```python
#converting the prediction probabilities into class label

Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```
